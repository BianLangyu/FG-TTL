```python
import json

# 两个 json 文件路径
path1 = "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/data/LLMTTA/OlympiadBench/OE_TO_maths_en_COMP/OE_TO_maths_en_COMP_formatted.json"
path2 = "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/data/LLMTTA/OlympiadBench/OE_TO_physics_en_COMP/OE_TO_physics_en_COMP_formatted.json"
output_path = "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/data/LLMTTA/OlympiadBench/OE_TO_maths_physics_en_COMP_formatted.json"

# 读取两个 json 文件
with open(path1, "r", encoding="utf-8") as f1:
    data1 = json.load(f1)
with open(path2, "r", encoding="utf-8") as f2:
    data2 = json.load(f2)

# 合并数据（假设都是列表）
merged_data = data1 + data2

# 保存为新的 json 文件
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(merged_data, f, ensure_ascii=False, indent=2)
```
