```python
import re

def extract_and_average_extra_memory(log_file_path):
    extra_memory_values = []
    
    # 正则表达式：匹配“训练额外消耗显存（峰值 - 初始化后）: X.XX GB”
    pattern = r"训练额外消耗显存（峰值 - 初始化后）:\s*([0-9]+\.?[0-9]*)\s*GB"
    
    with open(log_file_path, 'r', encoding='utf-8') as file:
        for line in file:
            match = re.search(pattern, line)
            if match:
                value = float(match.group(1))
                extra_memory_values.append(value)
    
    if extra_memory_values:
        average = sum(extra_memory_values) / len(extra_memory_values)
        # print(f"提取到的额外消耗显存值: {extra_memory_values}")
        print("统计额外显存消耗（峰值 - 初始化后）:")
        print(f"平均值: {average:.2f} GB，最大值：{max(extra_memory_values):.2f} GB")
        return average
    else:
        print("未找到匹配的显存消耗记录。")
        return None
    
def extract_and_average_extra_time(log_file_path):
    extra_memory_values = []
    
    # 正则表达式：匹配“训练额外消耗显存（峰值 - 初始化后）: X.XX GB”
    pattern = r"训练耗时:\s*([0-9]+\.?[0-9]*)\s*秒"
    
    with open(log_file_path, 'r', encoding='utf-8') as file:
        for line in file:
            match = re.search(pattern, line)
            if match:
                value = float(match.group(1))
                if value > 100:  value = 0
                extra_memory_values.append(value/8)
    
    if extra_memory_values:
        average = sum(extra_memory_values) / len(extra_memory_values)
        # print(f"提取到的训练耗时值: {extra_memory_values}")
        print("统计训练耗时:")
        print(f"平均值: {average:.2f} 秒，最大值：{max(extra_memory_values):.2f} 秒")
        return average
    else:
        print("未找到匹配的显存消耗记录。")
        return None

# 使用示例
log_path = '/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_tent/gsm8k_test_formatted/naive_entropy/lr_5e-6-all_tokens-seed_42/memory_log.txt'  # 替换为你的实际日志文件路径
log_path = '/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_eata/gsm8k_test_formatted/naive_entropy/lr_5e-6-all_tokens-seed_42/memory_log.txt'
log_path = '/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_ttl/gsm8k_test_formatted/naive_entropy/lr_5e-6-all_tokens-seed_42/memory_log.txt'
log_path = '/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_adaptive_k/relative_entropy_as_thres-without_abs/gsm8k_test_formatted/SAFE_Entropy_loss_detach_v2/lam_0.5-lr_7.5e-6-new_tokens_32-N0_8-K_max_24-K_min_2-initial_delta_0.02-patient_m_2-threshold_low_0.02-coef_0.3-seed_42/memory_log.txt'
paths = [
    # "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_tent/math_500/naive_entropy/lr_5e-6-all_tokens-seed_42/memory_log.txt",
    # "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_tent/math_500/dirichlet_entropy/lr_5e-6-all_tokens-seed_42/memory_log.txt",
    # "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_eata/math_500/naive_entropy/lr_5e-6-all_tokens-seed_42/memory_log.txt",
    # "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_ttl/math_500/naive_entropy/lr_5e-6-all_tokens-seed_42/memory_log.txt",
    # "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_adaptive_k/relative_entropy_as_thres-without_abs/math_500/SAFE_Entropy_loss_detach_v2/lam_0.5-lr_7.5e-6-new_tokens_32-N0_8-K_max_24-K_min_2-initial_delta_0.02-patient_m_2-threshold_low_0.02-coef_0.3-seed_42/memory_log.txt"
    # "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_tent/gsm8k_test_formatted/naive_entropy/lr_5e-6-all_tokens-seed_42-v2/memory_log.txt",
    # "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_eata/gsm8k_test_formatted/naive_entropy/lr_5e-6-all_tokens-seed_42-v2/memory_log.txt",
    # "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_tent/gsm8k_test_formatted/dirichlet_entropy/lr_5e-6-all_tokens-seed_42-v2/memory_log.txt",
    # "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_ttl/gsm8k_test_formatted/naive_entropy/lr_5e-6-all_tokens-seed_42-v2/memory_log.txt",
    # "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_adaptive_k/relative_entropy_as_thres-without_abs/gsm8k_test_formatted/SAFE_Entropy_loss_detach_v2/lam_0.5-lr_7.5e-6-new_tokens_32-N0_8-K_max_24-K_min_2-initial_delta_0.02-patient_m_2-threshold_low_0.02-coef_0.3-seed_42-v2/memory_log.txt"

    "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_tent/college_math_formatted/naive_entropy/lr_5e-6-all_tokens-seed_42/memory_log.txt",
    "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_eata/college_math_formatted/naive_entropy/lr_5e-6-all_tokens-seed_42/memory_log.txt",
    "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_tent/college_math_formatted/dirichlet_entropy/lr_5e-6-all_tokens-seed_42/memory_log.txt",
    "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_ttl/college_math_formatted/naive_entropy/lr_5e-6-all_tokens-seed_42/memory_log.txt",
    "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/record_time_and_mem/online_adaptive_k/relative_entropy_as_thres-without_abs/college_math_formatted/SAFE_Entropy_loss_detach_v2/lam_0.5-lr_7.5e-6-new_tokens_32-N0_8-K_max_24-K_min_2-initial_delta_0.02-patient_m_2-threshold_low_0.02-coef_0.3-seed_42/memory_log.txt"

]
for path in paths:
    extract_and_average_extra_memory(path)
    extract_and_average_extra_time(path)
    print("#"*50)
```
