```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 30

words = ['so', 'but', 'wait', 'therefore', 'maybe', 'alternatively']
y = [6407, 2358, 2014, 955, 877, 477]
freq = [i / 200 for i in y]

# 创建纵向柱状图
fig, ax = plt.subplots(figsize=(8, 6))
x_pos = np.arange(len(words))


# 绘制纵向柱状图
bars = ax.bar(x_pos, freq, color="#de7a60")

# 设置 X 轴标签倾斜
ax.tick_params(axis='both', labelsize=28)
ax.set_xticks(x_pos)
ax.set_xticklabels(words, rotation=20)  # 刻度标签倾斜 30 度

# 设置 Y 轴范围和网格
ax.set_ylim(0, max(freq) + 4)  # 根据数据调整上限
ax.yaxis.grid(True, alpha=0.6)

# 设置标签
ax.set_ylabel('Frequency') 
# ax.set_xlabel('Connection Words', fontsize=14)
# ax.set_title('Word Frequencies in Reasoning', fontsize=16)

# 添加数值标签
for i, v in enumerate(freq):
    ax.text(i, v + 0.02, f'{v:.1f}', ha='center', va='bottom', fontsize=20)

plt.tight_layout()
# plt.savefig('/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory/saves/NIPS25_Images/words_frequencies_v3.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
