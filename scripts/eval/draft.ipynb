```python
import re
import matplotlib.pyplot as plt

# 正则表达式用于提取 optim_K 和 损失值
pattern = r"optim_K是:\s*(\d+), 损失值是：\s*([0-9\.eE+-]+)"

def parse_log_file(file_path):
    """读取并解析单个 log 文件"""
    optim_ks = []
    losses = []

    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            match = re.search(pattern, line)
            if match:
                optim_k = int(match.group(1))
                loss = float(match.group(2))
                optim_ks.append(optim_k)
                losses.append(loss)

    return optim_ks, losses

def plot_multiple_logs(file_paths, labels=None):
    """读取多个 log 文件，并绘图对比 optim_K 和 Loss"""
    plt.figure(figsize=(12, 6))

    # 子图 1: Optim K 对比
    ax1 = plt.subplot(2, 1, 1)
    for i, file_path in enumerate(file_paths):
        label = labels[i] if labels else f"File {i + 1}"
        optim_ks, _ = parse_log_file(file_path)
        ax1.plot(optim_ks, linestyle="-", label=label)
    ax1.set_title("Optim K Comparison")
    ax1.set_ylabel("Optim K")
    ax1.grid(True)
    ax1.legend()

    # 子图 2: Loss 对比
    ax2 = plt.subplot(2, 1, 2)
    for i, file_path in enumerate(file_paths):
        label = labels[i] if labels else f"File {i + 1}"
        _, losses = parse_log_file(file_path)
        ax2.plot(losses, linestyle="-", label=label)
    ax2.set_title("Loss Comparison")
    ax2.set_xlabel("Sample Index")
    ax2.set_ylabel("Loss Value")
    ax2.grid(True)
    ax2.legend()

    plt.tight_layout()
    plt.show()

# === 使用示例 ===
if __name__ == "__main__":
    # 多个 log 文件路径（请替换为你自己的路径）
    file_path1 = "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/online_adaptive_k/relative_entropy_as_thres-without_abs/gsm8k_test_formatted/first_naive-SAFE_Entropy_loss_detach/lr_5e-6-new_tokens_32-N0_8-K_max_24-K_min_2-initial_delta_0.02-patient_m_2-threshold_low_0.02-coef_0.3-seed_42/logfile.txt"
    file_path2 = "/hujinwu/wyf/projects/zhangzitian/projects/LLaMA-Factory-1cfe429/saves/AAAI26/Llama-3.1-8B-Instruct/online_adaptive_k/relative_entropy_as_thres-without_abs/gsm8k_test_formatted/naive_entropy/lr_5e-6-new_tokens_32-N0_8-K_max_24-K_min_2-initial_delta_0.02-patient_m_2-threshold_low_0.02-coef_0.3-seed_42/logfile.txt"

    log_files = [
        file_path1,
        file_path2
    ]

    # 可选：给每个文件添加标签
    labels = ["adaptive_k + SAFE", "adaptive_k"]

    # 绘图
    plot_multiple_logs(log_files, labels=labels)
```
